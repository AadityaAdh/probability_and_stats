### Central Limit theorem

The Central Limit Theorem (CLT) is a fundamental concept in statistics that explains how the distribution of sample means becomes approximately normal as the sample size grows, regardless of the shape of the population distribution. In other words, if you take many random samples from any population (whether it’s uniform, skewed, or otherwise) and calculate the average (mean) of each sample, the distribution of these sample means will approach a normal (bell-shaped) distribution as the sample size increases. This allows us to make inferences about population parameters even if we don’t know the original distribution.


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
each_sample_size=20000
```

```python
#aaba sample haru nikali
sample_mean_array=[]
for sample_num in range(20000):
  my_sample=np.random.uniform(low=1,high=2000,size=each_sample_size)
  my_mean=np.mean(my_sample)
  sample_mean_array.append(my_mean)
```

```python
sample_mean_array
```

```python
#mean ko distribution chai normal hunxa vanae ko
```

```python
plt.hist(sample_mean_array)
```

# using chi square

```python

sample_mean_array=[]
for sample_num in range(20000):
  my_sample=np.random.chisquare(5,size=each_sample_size)
  my_mean=np.mean(my_sample)
  sample_mean_array.append(my_mean)
```
